```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
users=pd.read_excel(r'C:\Users\gayat\Desktop\new projects\Enterprise Database Performance Monitoring System\Dataset\users_dataset.xlsx')
```

```python
# data description
users.info()
users.head()
users.columns

users.isnull().sum()
  
  
```

```python
#cleaning department column
users["department"]=users["department"].fillna("Unknown")

users.isnull().sum()

users["department"].unique()

users["department"]=users["department"].str.strip()
users["department"] = users["department"].str.lower()
users["department"].unique()

users["department"]=users["department"].replace({
  "hr":"HR",
  "human resources":"HR",
  
  "sales":"Sales",
  
  "engineering":"Engineering",
  "eng":"Engineering",
  
  "customer success":"Customer Success",
  "finance":"Finance",
  "unknown":"Unknown",
  
  "it support":"IT support",
  "information technology":"IT support",
  "it":"IT support",
  "it dept":"IT support",
  
  "marketing":"Marketing",
  "mktg":"Marketing",
  
  "operations":"Operations",
  "legal":"Legal",
  "product":"Product",
  
})

users["department"].unique()
```

```python
#duplicate data
users.duplicated().sum()
```

```python
#userid cleanining
users["user_id"].dtype
users["user_id"].isnull().sum()
users["user_id"].duplicated().sum()
users[users["user_id"] <= 0]
```

```python
#user name cleaning
users["username"].isnull().sum()
users["username"]=users["username"].str.strip()
users["username"]=users["username"].str.lower()
users["username"].nunique()
users["username"].duplicated().sum()
users[users["username"].duplicated(keep=False)]

duplicates=users[users.duplicated(subset=["username","department"],keep=False)]
duplicates.sort_values(["username","department"])
users=users.drop_duplicates(subset=["username","department"])
users.duplicated(subset=["username","department"]).sum()
```
